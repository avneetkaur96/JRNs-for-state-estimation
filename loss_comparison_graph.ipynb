```python
import matplotlib.pyplot as plt
import re

loss_data = []
val_loss_data = []

with open("down_pendulum_loss.txt", "r") as f:
    lines = f.readlines()
    for line in lines:
        loss_match = re.search(r"Loss:\s*([0-9.eE+-]+)", line)
        if loss_match:
            loss_data.append(float(loss_match.group(1)))
        
        val_match = re.search(r"Val Loss:\s*([0-9.eE+-]+)", line)
        if val_match:
            val_loss_data.append(float(val_match.group(1)))

# Plot
plt.figure(figsize=(10, 5))
plt.plot(loss_data, label="Training Loss")
plt.plot(val_loss_data, label="Validation Loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.savefig("dp_loss_comparison.png")
plt.savefig("dp_loss_comparison.svg")
plt.show()
```
